```python
#Basic MarksSheet

Eng =   int(input("English Marks          : "))
Urdu =  int(input("Urdu Marks             : "))
Math =  int(input("Maths Marks            : "))
Pst =   int(input("Pakistan Studies Marks : "))
Chem =  int(input("Chemistry Marks        : "))

total_marks = 500
obt_marks = (Eng+Urdu+Math+Pst+Chem)
percentage = (obt_marks/total_marks)*100

print("Total Marks    = " , total_marks)
print("Obtained Marks = " , obt_marks)
print("Percentage is  = " , percentage)

if percentage >=80:
  print("You have passed with Grade : A-1")
  print("Remarks >>> Excellent")

elif percentage >=70 and percentage <80:
    print("You have passed with Grade : A")
    print("Remarks >>> Very Good")

elif percentage >=60 and percentage <70:
      print("You have passed with Grade : B")
      Print("Remarks >>> Good")

elif percentage >=50 and percentage <60:
      print("You have passed with Grade : C")
      print("Remarks >>> Average")

else:
    print("Sorry ! You have Failed ")
```
